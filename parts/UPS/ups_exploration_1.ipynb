```python
df1 = df[(df.load_cap_w<1000)&(df.avg>0)&(df.full_load>0)]
plt.figure(figsize=(12,8))
plt.scatter(df1.full_load, df1.avg, c=df1.load_cap_w, cmap="Blues", s=75)
plt.colorbar(label='Load Capacity (watts)')
plt.axis([0,20,0,1000])
plt.title('Backup/Run Time under full load (minutes) vs. Price', fontsize=14)
plt.xlabel('Backup/Run Time under full load (minutes)', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.savefig(os.path.expanduser('~/Documents/GitHub/briancaffey.github.io/img/ups/full_load_vs_price.png'))
```
